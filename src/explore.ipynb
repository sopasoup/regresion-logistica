```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn as sk
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/4GeeksAcademy/logistic-regression-project-tutorial/main/bank-marketing-campaign-data.csv", sep = ";")
df.info()
```

```python
df = df.drop_duplicates().reset_index(drop=True)
df.isnull().sum()
```

```python
df.info()
```

```python
for column in df.columns:
    if df[column].dtype == 'object':
        df[column+'_n'] = pd.factorize(df[column])[0]

```

```python
variables_num= []
for column in df.columns:
    if df[column].dtype != 'object':
        variables_num.append(column)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
features_scaler = scaler.fit_transform(df[variables_num])
df_scaler = pd.DataFrame(features_scaler, index = df.index, columns = variables_num)
df_scaler
```

```python
from sklearn.model_selection import train_test_split

X = df_scaler.drop("y", axis = 1)
y = df_scaler["y"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state = 42)
```

```python
from sklearn.feature_selection import chi2, SelectKBest

selection_model = SelectKBest(chi2, k = 5)
selection_model.fit(X_train, y_train)
ix = selection_model.get_support()
X_train_sel = pd.DataFrame(selection_model.transform(X_train), columns = X_train.columns.values[ix])
X_test_sel = pd.DataFrame(selection_model.transform(X_test), columns = X_test.columns.values[ix])

X_train_sel.head()
```

```python
X_train_sel["y_n"] = list(y_train)
X_test_sel["y_n"] = list(y_test)
X_train_sel.to_csv("../data/processed/clean_train.csv", index = False)
X_test_sel.to_csv("../data/processed/clean_test.csv", index = False)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```

```python
from sklearn.model_selection import GridSearchCV

hyperparams = {
    "C": [0.001, 0.01, 0.1, 1, 10, 100, 1000],
    "penalty": ["l1", "l2", "elasticnet", None],
    "solver": ["newton-cg", "lbfgs", "liblinear", "sag", "saga"]
}

grid = GridSearchCV(model, hyperparams, scoring = "accuracy", cv = 5)
grid
```

```python
def warn(*args, **kwargs):
    pass
import warnings
warnings.warn = warn

grid.fit(X_train_sel, y_train)

print(f"Best hyperparameters: {grid.best_params_}")
```

```python
best_model = grid.best_estimator_
y_best_pred = best_model.predict(X_test_sel)
```

```python
accuracy_score(y_test,y_best_pred)
```
